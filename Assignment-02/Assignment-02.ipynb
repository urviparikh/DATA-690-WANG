### User inputs 10 integers 

```
my_intlist =[] #creating an empty list
MAX_TRY = 11 # this will ensure the list has 10 values

for i in range(1, MAX_TRY):

    while True:
        try:
            user_input = input("Please enter an integer:") # this will only allow integer
            in_input = int(user_input)
            break
        except: #error message will appear if integer not entered
            print("Please retry (only integers will be accepted")
            continue



    my_intlist.append(user_input)   # this will allow the integers to be added to the list
    print(f"You have entered #{i}:", user_input)

print("In conclusion, you have entered:", my_intlist)
```

### List of Integers

```
onlyint_list =[]
for user_in in my_intlist:
    x = int(user_in)
    onlyint_list.append(x)

onlyint_list # displays the integer list 
```

### Calculate Max and Min

```
# Below will be used to calculate max and min 
def minmax(onlyint_list, minimum=None, maximum=None):

    head, *tail = onlyint_list

    if minimum is None:
        minimum = [head]
    elif head < minimum[-1]:
        minimum.append(head)

    if maximum is None:
        maximum = [head]
    elif head > maximum[-1]:
        maximum.append(head)

    if tail:
        return minmax(tail, minimum, maximum)

    return minimum.pop(), maximum.pop()

if __name__ == "__main__":

    
    minimum, maximum = minmax(onlyint_list)
    print("The minimum is", minimum)
    print("The maximum is", maximum)
    
```

### Calculate Range of the list of integers

```
# the range is calculated by taking the maximum and subtracting from the minimum

print("The range is", maximum-minimum)
```

### Calculate Mean


```
# the mean will be calculated by first using Sum of the total numbers
# the sum of total will be divided by the length of the list in order to obtain mean

n=len(onlyint_list)

get_sum = sum(onlyint_list)
my_mean = get_sum / n


x= int(my_mean)
type(x)
print("The mean is",x) 
```

### Calculate Variance

```
# to calculate variance from previously calculated mean
m = my_mean
my_var = sum((xi - m) ** 2 for xi in onlyint_list) / len(onlyint_list)

y= int(my_var)
type(y)
print("The variance is", y)
```

### Calculate Standard Deviation

```
# to get standard deviation from previously calculated mean
my_sDev = my_mean**(1.0/2)

z=int(my_sDev)
type(z)
print("The Standard deviation is",z)
```
